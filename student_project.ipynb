```python
from tkinter import *
from tkinter import messagebox
import sqlite3
count=0
tot_rec=0
def db_init():
    global tot_rec
    global count
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    tblcmd = 'create table student_info (rn char(8), name char(40), br char(3), cgpa float(6,3))'
    try:
        curs.execute(tblcmd)
    except:
        ins_rec=curs.execute('select * from student_info')
        rec = curs.fetchall()
        tot_rec = len(rec)
    conn.commit()
    conn.close()
    if(tot_rec>0):
        count=1
        firstRec()
def firstRec():
    global count
    if(count==0):
        messagebox.showerror("SQL ERROR", "No Records...")
        return
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('select * from student_info')
    first = curs.fetchone()
    rn.set(first[0])
    name.set(first[1])
    branch.set(first[2])
    cgpa.set(first[3])
    conn.close()
    count = 1
    l5.configure(text=str(count)+"/"+str(tot_rec))
def preRec():
    global count
    global tot_rec
    if(count==0):
        messagebox.showerror("SQL ERROR", "No Records...")
        return        
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('select * from student_info')
    if(count==1):
        first = curs.fetchone()
    else:
        for i in range(1,count):
            first = curs.fetchone()
        count -= 1
    rn.set(first[0])
    name.set(first[1])
    branch.set(first[2])
    cgpa.set(first[3])
    conn.close()
    l5.configure(text=str(count)+"/"+str(tot_rec))
def nextRec():
    global count
    global tot_rec
    if(count==0):
        messagebox.showerror("SQL ERROR", "No Records...")
        return
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('select * from student_info')
    if(tot_rec == 1):
        first = curs.fetchone()
    else:
        for i in range(1,count+1):
            first = curs.fetchone()
    if(count<tot_rec):
        first = curs.fetchone()
    else:
        count -= 1
    rn.set(first[0])
    name.set(first[1])
    branch.set(first[2])
    cgpa.set(first[3])
    conn.close()
    count += 1
    l5.configure(text=str(count)+"/"+str(tot_rec))
def lastRec():
    global count
    global tot_rec
    if(count==0):
        messagebox.showerror("SQL ERROR", "No Records...")
        return
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('select * from student_info')
    if(tot_rec==1):
        first = curs.fetchone()
    else:
        count = 0
        for i in range(1,tot_rec+1):
            first = curs.fetchone()
            count += 1
    rn.set(first[0])
    name.set(first[1])
    branch.set(first[2])
    cgpa.set(first[3])
    conn.close()
    l5.configure(text=str(count)+"/"+str(tot_rec))
def addRec():
    global tot_rec
    global count
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('insert into student_info values (?, ?, ?, ?)', (rn.get(), name.get(), branch.get(), eval(cgpa.get())))
    conn.commit()
    conn.close()
    tot_rec += 1
    count = tot_rec
    l5.configure(text=str(count)+"/"+str(tot_rec))
def delRec():
    global count
    global tot_rec
    if(count==0):
        messagebox.showerror("DELETE ERROR", "No Records...")
        return
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('delete from student_info where rn=?',(rn.get(),))
    conn.commit()
    conn.close()
    count = 1
    tot_rec -= 1
    firstRec()
def updateRec():
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('update student_info set name=?,br=?,cgpa=? where rn=?',(name.get(),branch.get(),cgpa.get(),rn.get()))
    conn.commit()
    conn.close()
    count = 1
    firstRec()
def searchRec():
    conn = sqlite3.connect('studentDB')
    curs = conn.cursor()
    ins_rec=curs.execute('select * from student_info where rn like ?',(rn.get(),))
    first = curs.fetchone()
    rn.set(first[0])
    name.set(first[1])
    branch.set(first[2])
    cgpa.set(first[3])
    conn.close()

win = Tk()
win.title("Student Information System")
win.geometry("700x300")
l1 = Label(win,text="Roll No.",font=("cambria",15,"bold"),bg="blue",fg="white", width=8, anchor="w")
l1.grid(row=1,column=1)
rn = StringVar()
e1 = Entry(win,textvariable = rn,width = 45,font=("cambria",15,"bold"),bg="orange")
e1.grid(row=1,column=2,columnspan=3)
l2 = Label(win,text="Name",font=("cambria",15,"bold"),bg="blue",fg="white", width=8, anchor="w")
l2.grid(row=2,column=1)
name = StringVar()
e2 = Entry(win,textvariable = name,width = 45,font=("cambria",15,"bold"),bg="orange")
e2.grid(row=2,column=2,columnspan=3)
l3 = Label(win,text="Branch",font=("cambria",15,"bold"),bg="blue",fg="white", width=8, anchor="w")
l3.grid(row=3,column=1)
branch = StringVar()
e3 = Entry(win,textvariable = branch,width = 45,font=("cambria",15,"bold"),bg="orange")
e3.grid(row=3,column=2,columnspan=3)
l4 = Label(win,text="CGPA",font=("cambria",15,"bold"),bg="blue",fg="white", width=8, anchor="w")
l4.grid(row=4,column=1)
cgpa = StringVar()
e4 = Entry(win,textvariable = cgpa,width = 45,font=("cambria",15,"bold"),bg="orange")
e4.grid(row=4,column=2,columnspan=3)
b1 = Button(win, width="8", text="|<", command=firstRec,font=("cambria",15,"bold"))
b1.grid(row=5, column=1,padx=10,pady=10)
b2 = Button(win, width="8", text="<", command=preRec,font=("cambria",15,"bold"))
b2.grid(row=5, column=2)
b3 = Button(win, width="8", text=">", command=nextRec,font=("cambria",15,"bold"))
b3.grid(row=5, column=3)
b4 = Button(win, width="8", text=">|", command=lastRec,font=("cambria",15,"bold"))
b4.grid(row=5, column=4)
b5 = Button(win,width="10",text="ADD", command=addRec,font=("cambria",15,"bold"))
b5.grid(row=6,column=1,padx=10,pady=10)
b6 = Button(win,width="10",text="DELETE", command=delRec,font=("cambria",15,"bold"))
b6.grid(row=6,column=2)
b7 = Button(win,width="10",text="UPDATE", command=updateRec,font=("cambria",15,"bold"))
b7.grid(row=6,column=3)
b8 = Button(win,width="15",text="SEARCH Roll No.", command=searchRec,font=("cambria",15,"bold"))
b8.grid(row=6,column=4)
l5 = Label(win,text=str(count)+"/"+str(tot_rec),font=("cambria",15,"bold"),bg="white", width=8, anchor="c")
l5.grid(row=7,column=1,columnspan=4)
db_init()
l5.configure(text=str(count)+"/"+str(tot_rec))
win.mainloop()
```

```python
def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n

        m = oldn
        n = oldm%oldn
    return n

class Fraction:
     def __init__(self,top,bottom):
         self.num = top
         self.den = bottom

     def __str__(self):
         return str(self.num)+"/"+str(self.den)

     def show(self):
         print(self.num,"/",self.den)

     def __add__(self,otherfraction):
         newnum = self.num*otherfraction.den + \
                      self.den*otherfraction.num
         newden = self.den * otherfraction.den
         common = gcd(newnum,newden)
         return Fraction(newnum//common,newden//common)
    
    def __mul__(self,otherfraction):
         newnum = self.num*otherfraction.den + \
                      self.den*otherfraction.num
         newden = self.den * otherfraction.den
         common = gcd(newnum,newden)
         return Fraction(newnum//common,newden//common)

     def __eq__(self, other):
         firstnum = self.num * other.den
         secondnum = other.num * self.den
         return firstnum == secondnum

a = list(map(int,input().split()))
b = list(map(int,input().split()))
x = Fraction(a[0],a[1])
y = Fraction(b[0],b[1])
print(x+y)
print(x == y)
```

```python
def gcd(n, d):
    while n != d:
        if n > d:
            n = n - d
        else:
            d = d - n
    return n
class Fraction:
    def __init__(self, n, d):
        self.num = int(n / gcd(abs(n), abs(d)))
        self.denom = int(d / gcd(abs(n), abs(d)))
        if self.denom < 0:
            self.denom = abs(self.denom)
            self.num = -1*self.num
        elif self.denom == 0:
            raise ZeroDivisionError
        
    def Add(self, other):
        return self.num*other.denom + self.denom*other.num, self.denom*other.denom
    def Mul(self, other):
        return self.num*other.num, self.denom*other.denom

def main():
    a = list(map(int,input().split()))
    b = list(map(int,input().split()))
    x = Fraction(a[0],a[1])
    y = Fraction(b[0],b[1]) 
    z = int(input())
    if (z == 1):
        q = list(Fraction.Add(x,y))
        print(str(q[0])+"/"+str(q[1]))
    elif (z==2):
        q = Fraction.Mul(x,y)
        print(str(q[0])+"/"+str(q[1]))
    
if __name__ == '__main__':
    main()
```
