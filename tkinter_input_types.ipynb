```python
from tkinter import * #package
from tkinter import ttk
win = Tk() #event loop
# win.geometry("500x400")
win.title("Python GUI") #created an object

label_a = Label(win, text = "COMBO BOX")
label_a.grid(column = 0, row = 0)

chVarDis = IntVar()
check1 = Checkbutton(win, text = "DISABLED", variable = chVarDis, state='disabled')
check1.grid(column = 0, row = 2, sticky=W) #tk.W
check1.deselect()
chVarUn = IntVar()
check2 = Checkbutton(win, text = "UnChecked", variable = chVarUn)
check2.deselect()
check2.grid(column = 0, row = 3, sticky=W) #tk.W
chVarEn = IntVar()
check3 = Checkbutton(win, text = "Enabled", variable = chVarEn)
check3.select()
check3.grid(column = 0, row = 4, sticky=W) #tk.W

win.mainloop() 
```

```python
from tkinter import * #package
from tkinter import ttk
win = Tk() #event loop
# win.geometry("500x400")
win.title("Python GUI") #created an object


COLOR1 = "#ff0000"
COLOR2 = "Green"
COLOR3 = "Blue"
def radCall():
    radSel = radVar.get()
    if radSel == 1:
        win.configure(background = COLOR1)
    elif radSel == 2:
        win.configure(background = COLOR2)
    elif radSel == 3:
        win.configure(background = COLOR3)

radVar = IntVar()
check1 = tk.Radiobutton(win, text = "COLOR 1", variable = radVar, value = 1, command = radCall)
check1.grid(column = 0, row = 2, sticky=W) #tk.W
check2 = tk.Radiobutton(win, text = "COLOR 2", variable = radVar, value = 2, command = radCall)
check2.grid(column = 0, row = 3, sticky=W) #tk.W
check3 = tk.Radiobutton(win, text = "COLOR 3", variable = radVar, value = 3, command = radCall)
check3.grid(column = 0, row = 4, sticky=W) #tk.W

win.mainloop() 
```

```python
from tkinter import * #package
from tkinter import ttk
win = Tk() #event loop
# win.geometry("500x400")
win.title("Python GUI") #created an object

label_name = Label(win, text = "NAME", height=2, width=20)
label_name.grid(column = 0, row = 0)
name = StringVar()
nameEntered = Entry(win, textvariable = name, width=20)
nameEntered.grid(column = 1, row = 0)

label_roll = Label(win, text = "ROLL NO.", height=2, width=20)
label_roll.grid(column = 0, row = 1)
roll = StringVar()
rollEntered = Entry(win, textvariable = roll, width=20)
rollEntered.grid(column = 1, row = 1)

label_marks = Label(win, text = "MARKS", height=2, width=20)
label_marks.grid(column = 0, row = 2)
marks = IntVar()
numberChosen = ttk.Combobox(win, textvariable = marks, state = "readonly", width=20)
x = []
for i in range(100):
    x.append(i)
numberChosen['values'] = tuple(x)
print(numberChosen.type())
numberChosen.grid(column = 1, row = 2)
numberChosen.current(0)

label_details = Label(win, text = "")
label_details.grid(column = 0, row = 3, columnspan = 2)
label_grade = Label(win, text = "")
label_grade.grid(column = 0, row = 4, columnspan = 2)

def clickMe():
    label_details.configure(text = "Student {0} with Roll No {1} has secured {2} marks".format(name.get(),roll.get(),marks.get()), foreground='red')
    grade = None
    if (marks.get()>0 and marks.get()<=20):
        grade = "D"
    elif (marks.get()>20 and marks.get()<=30):
        grade = "C+"
    elif (marks.get()>30 and marks.get()<=40):
        grade = "B"
    elif (marks.get()>40 and marks.get()<=50):
        grade = "B+"
    elif (marks.get()>50 and marks.get()<=70):
        grade = "A"
    elif (marks.get()>70 and marks.get()<=100):
        grade = "A+"
    label_grade.configure(text = "Student gets {0} grade.".format(grade), foreground='red')
action = Button(win, text = "CLICK ME", command = clickMe)
action.grid(column=1, row=5)

nameEntered.focus()
win.mainloop() 
```
