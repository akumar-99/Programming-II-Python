```python
from tkinter import * #package
from tkinter import ttk
win = Tk() #event loop
win.geometry("1024x720")
win.title("Python GUI") #created an object

label_name = Label(win, text = "Name of Customer", height=2)
label_name.grid(column = 0, row = 0)
name = StringVar()
nameEntered = Entry(win, textvariable = name)
nameEntered.grid(column = 1, row = 0)

label_bill = Label(win, text = "Bill Amount", height=2)
label_bill.grid(column = 0, row = 1)
bill = IntVar()
billEntered = Entry(win, textvariable = bill, width=20)
billEntered.grid(column = 1, row = 1)

label_payment = Label(win, text = "Mode of Payment", height=2, width=20)
label_payment.grid(column = 0, row = 2)
payment = StringVar()
paymentChosen = ttk.Combobox(win, textvariable = payment, state = "readonly", width=20)
paymentChosen['values'] = ('Cash', "Cheque", 'Credit Card')
paymentChosen.grid(column = 1, row = 2)
paymentChosen.current(0)

def calculateDiscount():
    billed = bill.get()
    discount1 = 0
    if (billed>=15000):
        discount1 = billed * 0.10
        billed = billed * 0.90
    mode = payment.get()
    if (mode == "Cash"):
        discount1 += billed * 0.08
    elif (mode == "Cheque"):
        discount1 += billed * 0.07
    discount.set(discount1)
    
action = Button(win, text = "Calculate Discount", command = calculateDiscount)
action.grid(column=0, row=3)

def calculateNetAmount():
    billed = bill.get()
    amounted = billed - discount.get()
    amount.set(amounted)

action = Button(win, text = "Calculate Net Amount", command = calculateNetAmount)
action.grid(column=1, row=3)

label_discount = Label(win, text = "Discount", height=2)
label_discount.grid(column = 0, row = 4)
discount = IntVar()
discountCalculated = Entry(win, textvariable = discount, width=20, state = 'disabled')
discountCalculated.grid(column = 1, row = 4)

label_amount = Label(win, text = "Net Amount", height=2)
label_amount.grid(column = 0, row = 5)
amount = IntVar()
amountCalculated = Entry(win, textvariable = amount, width=20, state = "disabled")
amountCalculated.grid(column = 1, row = 5)

Button(win, text="Quit", command=win.destroy).grid(column = 1, row = 6)



nameEntered.focus()
win.mainloop() 
```
